# 11.1

An analyst fits an RF classifier where some of the features include seasonally adjusted employment data. He aligns with January data the seasonally adjusted value of January, etc. What "sin" has he committed?

**A: January data is released in February and generally revised 3 times thereafter, i.e. the lookahead sin.**

# 11.2

An analyst develops an ML algorithm where he generates a signal using closing prices, and exectude at the close. What's the sin?

**A: Closing prices aren't known until close, i.e. the lookahead & transaction cost sins.**

# 11.3

There is a 98.51% correlation between total revenue generated by arcades and computer science doctorates awarded in the United States. As the number of doctorates is expected to grow, should we invest in arcades companies? If not, what's the sin?

**A: If there's no plausible way to explain why these are correlated, it's probably just a coincidence, i.e. the storytelling sin.**

# 11.4

The *Wall Street Journal* has reported that September is the only month of the year that has negative stock returns, looking back 20, 50, and 100 years. Should we sell stock at the end of August? If not, what's the sin?

**A: I don't think there's much of a sin here, seasonal patters (across all timeframes) such as these are very real in the markets.**

# 11.5

We download P/E ratios from Bloomberg, rank stocks every month, sell the top quartile, and buy the ~~long~~ bottom quartile. Performance is amazing. What's the sin?

**A: Knowing whether we would've been able to and how much it would've cost to short these issues is hard, i.e. shorting sin.**
